도커 허브에 
이미지로 판다스 파이선 c 뭐 다있음

도커 이미지 실행 docker run -d -p 80:80 <image_name>
터미널에서 부터 때어져라 -d -p 80:80 # 80 에서 80로 연결해라

443 = https
22 = SSH
