```python
# 1. 은행 계좌를 관리하는 클래스를 만들어보세요.

class BankAccount:
    def __init__(self, balance=0):
        self.__balance = balance  # private 속성

    def deposit(self, amount):
        self.__balance += amount
        
    def withdraw(self, amount):
        if self.__balance >= amount:
            self.__balance -= amount
        else:
            print('잔액이 부족합니다')
    
    def get_balance(self, id, pw):
        if id == "sad" and pw == "asdasd":
            return self.__balance
        else:
            print('계좌 비밀번호가 틀렸습니다') 

account = BankAccount(1000)
account.deposit(500)    
print(account.get_balance(input('id 입력하세요: '), input('비밀번호 입력하세요: ')))





```
