```python
"""
오늘 할 거 
객체지향 연습하고 o
try except 연습하고 o
if else 문 o
for while 문  o
with 문 연습 o
그 다음 다시 ai미연시 프로젝트 다시해
- 응 지금하고있음


--------------------
공식문서 보는게 짱짱맨이구나

리드미 
프로젝트 소개서



--------------------

sa 문서
기능에 대한 기술적인 부분=SA 문서



--------------------
"""
# 다형성 (Polymorphism)


# 같은 이름의 메서드가 서로 다르게 동작

# 도형 클래스
class  Shape:
    def area(delf):
        pass

# 사각형 클래스

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height
    

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2
    
shapes = [Rectangle(2, 3), Circle(3)]
for shape in shapes:
    print(shape.area())




```

```python
# 동물 기본 클래스
class Animal:
    def __init__(self, name):
        self.name = name
    
    def make_sound(self):
        pass
    
    def move(self):
        pass

# 강아지 클래스
class Dog(Animal):
    def make_sound(self):
        return f"{self.name}가 멍멍 짖습니다!"
    
    def move(self):
        return f"{self.name}가 네 발로 뛰어갑니다!"

# 새 클래스
class Bird(Animal):
    def make_sound(self):
        return f"{self.name}가 짹짹 웁니다!"
    
    def move(self):
        return f"{self.name}가 하늘을 납니다!"

# 물고기 클래스
class Fish(Animal):
    def make_sound(self):
        return f"{self.name}가 뽀글뽀글 거립니다!"
    
    def move(self):
        return f"{self.name}가 헤엄칩니다!"

# 다형성 활용 예제
def animal_chorus(animals):
    print("=== 동물들의 소리 ===")
    for animal in animals:
        print(animal.make_sound())
    
    print("\n=== 동물들의 움직임 ===")
    for animal in animals:
        print(animal.move())

# 실행 예시
my_animals = [
    Dog("멍멍이"),
    Bird("짹짹이"),
    Fish("뽀글이")
]

animal_chorus(my_animals)
```

```python
# with 문
with open("example.txt", "w") as file:
    file.write("Hello, world!\\n")
    file.write("This is a file example\\n")

with open("example.txt", "r") as file:
    content = file.read()
    print(content)
    
```

with 문은 Python에서 컨텍스트 매니저(Context Manager)를 사용하여 자원(파일, 네트워크 연결 등)의 획득과 해제를 자동으로 처리하는 구문입니다. with 문을 사용하면 코드 블록이 종료될 때 자동으로 자원을 정리할 수 있습니다.


```python
with open("example.txt", "r") as file:
    content = file.read()
    print(content)

# 이 예제에서 with 문을 사용하여 파일을 열고, 파일을 읽은 후 자동으로 파일을 닫습니다. 
# open 함수는 파일 객체를 반환하며,
# with 문은 파일 객체의 컨텍스트 매니저 역할을 합니다. 
# with 블록이 종료되면 파일이 자동으로 닫힙니다.
```
