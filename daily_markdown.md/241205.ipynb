# 학습 방법

1. 창을 분할하여 한쪽은 notion, 한쪽은 IDE 세팅
2. 해당 문법 정리 글을 보면서 동시에 **실습 수행(필수)**
    - Tips. 실습을 수행하실 때, notion에 나온 예시를 포함한 스스로 데이터를 생성해서 해당 문법을 적용해보세요! (실력이 매우 늘지도~?)
3. 특정 문법이 더 궁금하다면 아래와 같은 방법 수행
    
    3.1 튜터 상담
    
    3.2 구글 검색 창에 [특정 문법] or [특정 문법 예시] 검색 후 블로그 및 공식 문서 확인 or ChatGPT
    
    (ex. python replace, python replace 예시, python replace 사용법 …)

1. 절대 눈으로 읽고 지나가지 마세요!
