# 할거 다른 팀원들 프론트엔드 실행할수있게 가이드 써놓기

Node.js 공식 웹사이트에 방문하여 LTS 버전을 다운로드하고 설치 

---

설치 되었는지 명령 프롬프트 열고 확인 
node -v
npm -v 

---

클론을 하던지 다운을 받던지  frontend폴더 받고

git clone <repository-url>
cd <repository-directory>

---

터미널  my-diary-app폴더 위치에서 명령어 실행 하면 켜짐

my-diary-app% npm run dev
