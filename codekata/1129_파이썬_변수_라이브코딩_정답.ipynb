deep
- 정수형(int)은 불변(Immutable) 객체입니다. 즉, 값이 변경되지 않습니다.

### 문제 1번 DeepDive
y가 항상 최신화된 x의 값을 가지고 싶다면?  

1. 변경 가능한(Mutable) 객체 사용하기 -> 리스트나 딕셔너리
```python
x = [5]  # 리스트로 감싸서 변경 가능한 객체로 만듭니다.
y = x    # y와 x는 같은 리스트 객체를 가리킵니다.

x[0] = 10  # 리스트의 첫 번째 요소를 변경합니다.

print(y[0])  # 출력: 10
```

2. 클래스를 사용하여 속성 공유하기
```python
class Number:
    def __init__(self, value):
        self.value = value

x = Number(5)
y = x  # x와 y는 동일한 Number 객체를 가리킵니다.

x.value = 10  # 객체의 속성 value를 변경합니다.

print(y.value)  # 출력: 10
```

3. 함수를 사용하여 동적으로 값 가져오기
```python
x = 5

def y():
    return x

x = 10
print(y())  # 출력: 10
```

## 문제 10
전역 변수와 지역 변수 얼마나 알고 계신가요?

참고 자료 : https://www.notion.so/teamsparta/DeepDive-b09b7ef419104d8a8d7c483c9432f89b

변수의 범위(Scope)란 프로그램에서 변수가 유효하게 작동하는 영역을 의미합니다. 즉, 어떤 변수에 접근할 수 있는 코드의 부분을 나타냅니다. 파이썬에서는 변수의 범위가 코드의 구조(예: 함수, 클래스 등)에 따라 결정됩니다.

변수의 범위는 크게 두 가지로 나뉩니다:

- 전역 변수(Global Variable): 프로그램 전체에서 접근 가능한 변수.
- 지역 변수(Local Variable): 특정 블록(주로 함수나 메서드) 내에서만 접근 가능한 변수.

아래의 코드에서 전역 변수와 지역 변수를 찾아보세요!

````python
count = 0

def increment():
    global count
    count += 1
    
def greet():
    message = "안녕하세요"
    print(message)

greet() 
increment()

# count 는 전역변수 
# message 는 지역변수
````

아래의 출력 결과를 생각해보세요

````python
x = 'global x'  # 전역 변수 x 선언 -------- 1

def outer(): # -------- 3
    x = 'enclosed x'  # 외부 함수의 지역 변수 x 선언 -------- 4
    def inner(): # -------- 6
        x = 'local x'  # 내부 함수의 지역 변수 x 선언 -------- 7
        print(x)       # 내부 함수에서 x 출력 --------  8
    inner()            # 내부 함수 호출 -------- 5
    print(x)           # 외부 함수에서 x 출력 -------- 9

outer()                # 외부 함수 호출 -------- 2
print(x)               # 전역 범위에서 x 출력 --------  10

# 음 같은 라인에 따라 지역변수 우선순위가 결정된다.
````
