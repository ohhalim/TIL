:        square_number(num) .  
: square_number(5)  25  .

```python
def square_number(num):
    squared_number = num **2
    return squared_number

result = square_number(5)
print(result)
```

```python
def to_uppercase(s):
    return s.upper()

print(to_uppercase("hello"))
```

: max_of_two(5, 10) -> 10, max_of_two(-3, -1) -> -1

```python
def max_of_two(a, b):
    if a > b:
        return a
    else: 
        return b
```

##  6
:   ,          average(lst) .    0 .  
: average([10, 20, 30]) -> 20.0, average([]) -> 0

```python
def average(lst):
    if len(lst)== 0:
        return 0
    sum_num = sum(lst)
    len_num = len(lst)
    return sum_num/len_num
```

##  7
:    start, end  , start end()     sum_range(start, end) .  
start end  ,      .  
, sum_range(5, 3) 3 5  3+4+5=12  .  

: sum_range(1, 3) -> 6(1+2+3), sum_range(3, 1) -> 6(1+2+3)

```python
def sum_range(start, end):
    if start > end: 
        end = start
        start = end
    total =0
    for i in range(start, end+1):
        total += i
    return total
```
