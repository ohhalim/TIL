## 문제 7
1부터 100 사이의 임의의 정답 숫자를 정해놓고, 사용자가 숫자를 추측하도록 합니다.  
사용자가 입력한 숫자가 정답보다 크면 "너무 큽니다.", 작으면 "너무 작습니다."를 출력하고, 정답을 맞추면 "정답입니다!"를 출력한 뒤 반복을 종료하세요.  
(정답은 코드 내에 직접 지정하거나 난수로 고정)

```python
answer = 10

while True:
    guess = int(input("입력:"))
    if guess > answer:
        print("큽니다")

    elif guess < answer:
        print("작습니다")

    elif guess == answer:
        print("맞습니다")

        break


# else 문에는 아무것도 안 적는다 적으면 에러 나옴
```

## 문제 9
사용자로부터 정수를 입력받아, 해당 정수의 각 자릿수의 합을 구하는 프로그램을 작성하세요.  
예: 1234 → 1+2+3+4=10

```python
n = int(input("정수를 입력하세요:"))
sum = 0 
for i in str(n):
    sum += int(i)
print(f"{n} 의 각 자릿수의 합은 {sum}입니다")

'''
str(n): 정수 n을 문자열로 변환합니다. 예를 들어, n이 1234라면 str(n)은 "1234"가 됩니다.
for i in str(n): 문자열 "1234"의 각 문자를 반복합니다. 각 문자는 i에 저장됩니다.
int(i): 현재 문자를 정수로 변환합니다. 예를 들어, i가 "1"이라면 int(i)는 1이 됩니다.
sum += int(i): 현재 문자를 정수로 변환한 값을 sum에 더합니다. 이는 sum = sum + int(i)와 동일합니다.
f"{n} 의 각 자릿수의 합은 {sum}입니다": f-string을 사용하여 문자열을 포맷합니다. {n}과 {sum}은 각각 변수 n과 sum의 값을 문자열에 삽입합니다.
print(...): 포맷된 문자열을 출력합니다.
예제 실행
사용자가 1234를 입력하면 프로그램은 다음과 같이 동작합니다:

n에 1234가 저장됩니다.
sum 변수를 0으로 초기화합니다.
str(n)은 "1234"가 되고, 각 자릿수를 반복합니다:
i = "1"일 때, sum += int("1")는 sum = 0 + 1 = 1이 됩니다.
i = "2"일 때, sum += int("2")는 sum = 1 + 2 = 3이 됩니다.
i = "3"일 때, sum += int("3")는 sum = 3 + 3 = 6이 됩니다.
i = "4"일 때, sum += int("4")는 sum = 6 + 4 = 10이 됩니다.
결과를 출력합니다: 1234 의 각 자릿수의 합은 10입니다.
'''
```

```python
n = input(f'입력하세요:')
sum= 0
for i in n:
    sum += int(i)
print(sum)
```
