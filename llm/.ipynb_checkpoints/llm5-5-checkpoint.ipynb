```python
import os
from getpass import getpass

os.environ["OPENAI_API_KEY"] = getpass("OpenAI API key 입력: ")
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage
#모델 초기화 하는거라함
model = ChatOpenAI(model="gpt-4")
#모델에 메세지 전달 하는거라함 @.@
response = model.invoke([HumanMessage(content="안녕하세요, 무엇을 도와드릴까요?")])
#휴먼 메세지 인스턴스는 사람이 입력한 문자를 기계가 이해하게 만듬
print(response.content)
```

```python
from langchain_core.prompts import ChatPromptTemplate
# 코어라는 곳에서 탬플릿을 가져온다 아래쪽에서는 탬플릿을 정의하고있음 from 으로 객채를 생성 대화의 흐름을 설정 
# 인보크 저장하는 함수 @.@ 
#시그템 메세지 설정하는거라함 
system_template ="Translate the following sentence from English to {language}:"
#사용자 텍스트 입력하는거라함
prompt_template =ChatPromptTemplate.from_messages([
    ("system",system_template),
    ("user","{text}")
])

#프롬포트 생성
result = prompt_template.invoke({"language":"French", "text": "How are you?"})
print(result.to_messages())
```

```python
#랭체인 연결하는 거
from langchain_core.output_parsers import StrOutputParser
# 파서는 문자열형태로 바꾸는 걸함
# 기본적인 형태의 파서 텍스트로 변환
#
# 응답을 파싱하는 파서초기화
parser = StrOutputParser()

#템플릿, 모델 파서를 체인으로 연결
chain = prompt_template | model | parser
#체인 실행 
response = chain.invoke({"language":"Spanish","text":"Where is the library?"})
print(response)
```

```python
# openai 임베딩 모델로 백터 임베딩 생성
from langchain_openai import OpenAIEmbeddings
#openai 임베딩모델 초기화
embeddings = OpenAIEmbeddings(model="text-embedding-ada-002")
```

```python
#FAISS 인덱스 초기화
!pip install langchain-community

import faiss 
from langchain_community.vectorstores import FAISS
from langchain_community.docstore.in_memory import InMemoryDocstore

# FAISS 인덱스 생성
index = faiss.IndexFlatL2(len(embeddings.embed_query("hello world")))
vector_store = FAISS(
    embedding_function=embeddings,
    index=index,
    docstore=InMemoryDocstore(),
    index_to_docstore_id={}
)
```

```python
# 백터 베이스에 문서 추가하는 거

from langchain_core.documents import Document
from uuid import uuid4

# 문서 생성
documents = [
    Document(page_content="LangChain을 사용해 프로젝트를 구축하고 있습니다!", metadata={"source": "tweet"}),
    Document(page_content="내일 날씨는 맑고 따뜻할 예정입니다.", metadata={"source": "news"}),
    Document(page_content="오늘 아침에는 팬케이크와 계란을 먹었어요.", metadata={"source": "personal"}),
    Document(page_content="주식 시장이 경기 침체 우려로 하락 중입니다.", metadata={"source": "news"}),
]

# 고유 ID 생성 및 문서 추가
uuids = [str(uuid4()) for _ in range(len(documents))]
vector_store.add_documents(documents=documents, ids=uuids)
```

```python
# 백터 데에터베이스 쿼리
# 기본 유사성 검색
results = vector_store.similarity_search("내일 날씨는 어떨까요?", k=2, filter={"source": "news"})
for res in results:
    print(f"* {res.page_content} [{res.metadata}]")

# 점수와 함께 유사성 검색
results_with_scores = vector_store.similarity_search_with_score("LangChain에 대해 이야기해주세요.", k=2, filter={"source": "tweet"})
for res, score in results_with_scores:
    print(f"* [SIM={score:.3f}] {res.page_content} [{res.metadata}]")
```

```python
# Retriever로 변환
# FAISS를 `retriever`로 변환해 RAG 체인에서 사용합니다.
retriever = vector_store.as_retriever(search_type="similarity", search_kwargs={"k": 1})
```

```python
#RAG 체인 생성
#LangChain의 모델과 프롬프트를 연결하여 RAG 체인을 구성합니다.

from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnablePassthrough

# 프롬프트 템플릿 정의
contextual_prompt = ChatPromptTemplate.from_messages([
    ("system", "Answer the question using only the following context."),
    ("user", "Context: {context}\\n\\nQuestion: {question}")
])


class DebugPassThrough(RunnablePassthrough):
    def invoke(self, *args, **kwargs):
        output = super().invoke(*args, **kwargs)
        print("Debug Output:", output)
        return output
# 문서 리스트를 텍스트로 변환하는 단계 추가
class ContextToText(RunnablePassthrough):
    def invoke(self, inputs, config=None, **kwargs):  # config 인수 추가
        # context의 각 문서를 문자열로 결합
        context_text = "\n".join([doc.page_content for doc in inputs["context"]])
        return {"context": context_text, "question": inputs["question"]}

# RAG 체인에서 각 단계마다 DebugPassThrough 추가
rag_chain_debug = {
    "context": retriever,                    # 컨텍스트를 가져오는 retriever
    "question": DebugPassThrough()        # 사용자 질문이 그대로 전달되는지 확인하는 passthrough
}  | DebugPassThrough() | ContextToText()|   contextual_prompt | model

# 질문 실행 및 각 단계 출력 확인
response = rag_chain_debug.invoke("강사이름은?")
print("Final Response:")
print(response.content)
```

```python
## FAISS 인덱스의 저장 및 로드
#FAISS 인덱스를 저장해 다시 사용할 수 있습니다.
# 인덱스 저장
vector_store.save_local("faiss_index")

# 저장된 인덱스 로드
new_vector_store = FAISS.load_local("faiss_index", embeddings)
```

```python
#faiss 데이터 병합
db1 = FAISS.from_texts(["문서 1 내용"], embeddings)
db2 = FAISS.from_texts(["문서 2 내용"], embeddings)

# 병합
db1.merge_from(db2)
```

```python
"""
이 코드는 크게 다음과 같은 주요 부분으로 구성되어 있습니다:

초기 설정

OpenAI API 연결
모델 초기화
기본 대화 기능


번역 시스템

번역 템플릿 설정
체인 구성
번역 실행


벡터 데이터베이스

FAISS 초기화
문서 저장
유사도 검색


RAG 시스템

컨텍스트 기반 질답
디버깅 기능
문서 검색 및 응답 생성


데이터 관리

인덱스 저장/로드
데이터베이스 병합



특정 부분에 대해 더 자세한 설명이 필요하시다면 말씀해 주세요. CopyRetryClaude can make mistakes. Please double-check responses.
"""

# OpenAI API를 사용하기 위한 환경 변수 관련 모듈
import os
from getpass import getpass  # 비밀번호 안전하게 입력받기 위한 모듈

# API 키를 환경변수로 설정 (보안을 위해 직접 코드에 키를 작성하지 않음)
os.environ["OPENAI_API_KEY"] = getpass("OpenAI API key 입력: ")

# OpenAI의 ChatGPT 모델을 사용하기 위한 LangChain 인터페이스
from langchain_openai import ChatOpenAI

# 사용자 메시지를 형식화하기 위한 클래스
from langchain_core.messages import HumanMessage

# GPT-4 모델 초기화
model = ChatOpenAI(model="gpt-4")

# 모델에 메시지 전송하고 응답 받기
# HumanMessage: 사용자의 입력을 AI가 이해할 수 있는 형식으로 변환
response = model.invoke([HumanMessage(content="안녕하세요, 무엇을 도와드릴까요?")])

# AI의 응답 출력
print(response.content)

# 대화 템플릿을 위한 모듈 임포트
from langchain_core.prompts import ChatPromptTemplate

# 번역을 위한 시스템 프롬프트 정의
# {language}: 번역할 목표 언어를 위한 변수
system_template = "Translate the following sentence from English to {language}:"

# 대화 템플릿 생성
# from_messages: 메시지 리스트로부터 템플릿 생성
# ("system", system_template): 시스템 지시사항
# ("user", "{text}"): 사용자 입력을 위한 변수
prompt_template = ChatPromptTemplate.from_messages([
    ("system", system_template),
    ("user", "{text}")
])

# 템플릿에 실제 값을 넣어 프롬프트 생성
# language: "French", text: "How are you?"를 템플릿의 변수에 대입
result = prompt_template.invoke({"language":"French", "text": "How are you?"})
print(result.to_messages())

# 문자열 출력 파서 임포트 (AI 응답을 문자열로 변환)
from langchain_core.output_parsers import StrOutputParser

# 응답을 문자열로 변환하는 파서 초기화
parser = StrOutputParser()

# 템플릿, 모델, 파서를 하나의 체인으로 연결 (파이프라인 생성)
chain = prompt_template | model | parser

# 체인 실행 (스페인어로 번역 요청)
response = chain.invoke({"language":"Spanish", "text":"Where is the library?"})
print(response)

# OpenAI의 텍스트 임베딩(벡터화) 모델 임포트
from langchain_openai import OpenAIEmbeddings

# 임베딩 모델 초기화 (text-embedding-ada-002 모델 사용)
embeddings = OpenAIEmbeddings(model="text-embedding-ada-002")

# FAISS 벡터 데이터베이스 관련 모듈 설치 및 임포트
!pip install langchain-community
import faiss  # Facebook AI Similarity Search - 고성능 벡터 검색 라이브러리
from langchain_community.vectorstores import FAISS
from langchain_community.docstore.in_memory import InMemoryDocstore

# FAISS 인덱스 초기화
# embed_query("hello world")의 결과 길이로 벡터 차원 설정
index = faiss.IndexFlatL2(len(embeddings.embed_query("hello world")))

# 벡터 저장소 초기화
vector_store = FAISS(
    embedding_function=embeddings,  # 텍스트를 벡터로 변환하는 함수
    index=index,                   # FAISS 인덱스
    docstore=InMemoryDocstore(),   # 문서 저장소
    index_to_docstore_id={}        # 인덱스와 문서 ID 매핑
)

# 문서 처리를 위한 모듈들 임포트
from langchain_core.documents import Document
from uuid import uuid4  # 고유 식별자 생성

# 샘플 문서 생성
documents = [
    Document(page_content="LangChain을 사용해 프로젝트를 구축하고 있습니다!", metadata={"source": "tweet"}),
    Document(page_content="내일 날씨는 맑고 따뜻할 예정입니다.", metadata={"source": "news"}),
    Document(page_content="오늘 아침에는 팬케이크와 계란을 먹었어요.", metadata={"source": "personal"}),
    Document(page_content="주식 시장이 경기 침체 우려로 하락 중입니다.", metadata={"source": "news"}),
]

# 각 문서에 대한 고유 ID 생성
uuids = [str(uuid4()) for _ in range(len(documents))]

# 문서를 벡터 저장소에 추가
vector_store.add_documents(documents=documents, ids=uuids)

# 유사도 검색 실행
# k=2: 상위 2개 결과 반환
# filter: "news" 소스의 문서만 검색
results = vector_store.similarity_search("내일 날씨는 어떨까요?", k=2, filter={"source": "news"})
for res in results:
    print(f"* {res.page_content} [{res.metadata}]")

# 유사도 점수와 함께 검색 실행
results_with_scores = vector_store.similarity_search_with_score(
    "LangChain에 대해 이야기해주세요.", 
    k=2, 
    filter={"source": "tweet"}
)
for res, score in results_with_scores:
    print(f"* [SIM={score:.3f}] {res.page_content} [{res.metadata}]")

# 벡터 저장소를 검색기(retriever)로 변환
retriever = vector_store.as_retriever(
    search_type="similarity",  # 유사도 기반 검색
    search_kwargs={"k": 1}     # 상위 1개 결과만 반환
)

# RAG(Retrieval-Augmented Generation) 체인 생성을 위한 프롬프트 템플릿
contextual_prompt = ChatPromptTemplate.from_messages([
    ("system", "Answer the question using only the following context."),
    ("user", "Context: {context}\\n\\nQuestion: {question}")
])

# 디버깅을 위한 패스스루 클래스
class DebugPassThrough(RunnablePassthrough):
    def invoke(self, *args, **kwargs):
        output = super().invoke(*args, **kwargs)
        print("Debug Output:", output)
        return output

# 문서 리스트를 텍스트로 변환하는 클래스
class ContextToText(RunnablePassthrough):
    def invoke(self, inputs, config=None, **kwargs):
        context_text = "\n".join([doc.page_content for doc in inputs["context"]])
        return {"context": context_text, "question": inputs["question"]}

# RAG 체인 구성
rag_chain_debug = {
    "context": retriever,
    "question": DebugPassThrough()
} | DebugPassThrough() | ContextToText() | contextual_prompt | model

# 질문 실행
response = rag_chain_debug.invoke("강사이름은?")
print("Final Response:")
print(response.content)

# FAISS 인덱스 저장
vector_store.save_local("faiss_index")

# 저장된 인덱스 불러오기
new_vector_store = FAISS.load_local("faiss_index", embeddings)

# 데이터베이스 병합 예제
db1 = FAISS.from_texts(["문서 1 내용"], embeddings)
db2 = FAISS.from_texts(["문서 2 내용"], embeddings)
db1.merge_from(db2)  # db2를 db1에 병합
```
