```python
print(pages)
```

```python
'''
아래 코드는 문서를 청크 단위로 나누는 함수입니다.
'''
from langchain.text_splitter import CharacterTextSplitter

text_splitter = CharacterTextSplitter(
    separator="\n",
    chunk_size=100,
    chunk_overlap=20
)

chunks = text_splitter.split_documents(pages)
```
