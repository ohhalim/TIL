```python
# PyTorch    
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
```

```python
#  
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

# CIFAR-10  
trainset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=True)

testset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=64, shuffle=False)
```

```python
# CNN  
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 32, 3, padding=1)  #   3,   32,   3x3
        self.pool = nn.MaxPool2d(2, 2)               #   2x2
        self.conv2 = nn.Conv2d(32, 64, 3, padding=1) #   32,   64,   3x3
        self.fc1 = nn.Linear(64 * 8 * 8, 512)        #   
        self.fc2 = nn.Linear(512, 10)                #   (10 )

    def forward(self, x):
        x = self.pool(torch.relu(self.conv1(x)))
        x = self.pool(torch.relu(self.conv2(x)))
        x = x.view(-1, 64 * 8 * 8)  #  :             
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x

'''
- `nn.Conv2d`: 2   .
    - $\scriptsize\textsf{nn.Conv2d(in\_channels, out\_channels, kernel\_size, padding)   ,   ,  ,  }$
- `nn.MaxPool2d`: 2    .
    - $\scriptsize\textsf{nn.MaxPool2d(kernel\_size, stride)    .}$
- `view`:   .
    - $\scriptsize\textsf{x.view(-1, 64 * 8 * 8)   1  .}$
'''
```

```python
##  
#  
model = SimpleCNN()

#     
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.01, momentum=0.9)

#  
for epoch in range(10):  # 10   
    running_loss = 0.0
    for i, data in enumerate(trainloader, 0):
        inputs, labels = data

        #  
        optimizer.zero_grad()

        #  +  + 
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        #  
        running_loss += loss.item()
        if i % 100 == 99:  #  100  
            print(f'[Epoch {epoch + 1}, Batch {i + 1}] loss: {running_loss / 100:.3f}')
            running_loss = 0.0

print('Finished Training')

'''
- `nn.CrossEntropyLoss`:        .         .
- `optim.SGD`:   (Stochastic Gradient Descent)   .
    - $\scriptsize\textsf{lr ,momentum   .}$
- `optimizer.zero_grad()`:     .
- `loss.backward()`:    .
- `optimizer.step()`:     .
'''
```

```python
# 
correct = 0
total = 0
with torch.no_grad():
    for data in testloader:
        images, labels = data
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f'Accuracy of the network on the 10000 test images: {100 * correct / total:.2f}%')
'''
- `torch.no_grad()`:      ,     .
- `torch.max`:    .
    - $\scriptsize\textsf{torch.max(outputs.data, 1)         .}$
- `labels.size(0)`:   .
- `(predicted == labels).sum().item()`:        .
'''
```
